```python
import pandas as pd
```

```python
plants_info = {}
dfs1 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="WhCosts")
dfs2 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="WhCapacities")
dfs3 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="ProductsPerPlant")
dfs4 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="VmiCustomers")
dfs5 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="PlantPorts")

plant_cost_dict = dfs1.set_index('WH')['Cost/unit'].to_dict()
plant_capacity_dict = dfs2.set_index('Plant ID')['Daily Capacity '].to_dict()

for wh in plant_cost_dict.keys():
    plants_info[wh] = {"cost" : plant_cost_dict[wh]}

for wh in plant_capacity_dict.keys():
    plants_info[wh]["capacity"] = plant_capacity_dict[wh]
    plants_info[wh]["products"] = []
    plants_info[wh]["exceptions"] = []
    plants_info[wh]["ports"] = []

for index, row in dfs3.iterrows():
    wh = row['Plant Code']
    product = row['Product ID']  
    if wh in plants_info.keys():
        plants_info[wh]["products"].append(product)

for index, row in dfs4.iterrows():
    wh = row['Plant Code']
    customer = row['Customers']  
    plants_info[wh]["exceptions"].append(customer)

for index, row in dfs5.iterrows():
    wh = row['Plant Code']
    port = row['Port']  
    plants_info[wh]["ports"].append(port)
    

freight_info = {}
dfs6 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="FreightRates")

for index, row in dfs6.iterrows():
    carrier = row['Carrier']
    freight_info[carrier] = {'orig_port_cd': [],
                             'dest_port_cd': [],
                             'minm_wgh_qty': [],
                             'max_wgh_qty': [],
                             'svc_cd': [],
                             'minimum cost': [],
                             'rate': [],
                             'mode_dsc': [],
                             'tpt_day_cnt': [],
                             'Carrier type': [],
                            }
    
for index, row in dfs6.iterrows():
    carrier = row['Carrier']
    freight_info[carrier]['orig_port_cd'].append(row['orig_port_cd'])
    freight_info[carrier]['dest_port_cd'].append(row['dest_port_cd'])
    freight_info[carrier]['minm_wgh_qty'].append(row['minm_wgh_qty'])
    freight_info[carrier]['max_wgh_qty'].append(row['max_wgh_qty'])
    freight_info[carrier]['svc_cd'].append(row['svc_cd'])
    freight_info[carrier]['minimum cost'].append(row['minimum cost'])
    freight_info[carrier]['orig_port_cd'].append(row['orig_port_cd'])
    freight_info[carrier]['rate'].append(row['rate'])
    freight_info[carrier]['mode_dsc'].append(row['mode_dsc'])
    freight_info[carrier]['tpt_day_cnt'].append(row['tpt_day_cnt'])
    freight_info[carrier]['Carrier type'].append(row['Carrier type'])

    
order_info = {}
dfs7 = pd.read_excel('Supply chain logisitcs problem.xlsx', sheet_name="OrderList")

    
for index, row in dfs7.iterrows():
    id = row['Order ID']
    order_info[id] = {'Order Date': row['Order Date'],
                            'Origin Port': row['Origin Port'],
                            'Carrier': row['Carrier'],
                            'TPT': row['TPT'],
                            'Service Level': row['Service Level'],
                            'Ship ahead day count': row['Ship ahead day count'],
                            'Ship Late Day count': row['Ship Late Day count'],
                            'Customer': row['Customer'],
                            'Product ID': row['Product ID'],
                            'Plant Code': row['Plant Code'],
                            'Destination Port': row['Destination Port'],
                            'Unit quantity': row['Unit quantity'],
                            'Weight': row['Weight'],
                              }

print(order_info[1447398419.7])
```
